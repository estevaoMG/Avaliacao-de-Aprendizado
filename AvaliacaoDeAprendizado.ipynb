```python
# =========================================
# üìä C√°lculo de M√©tricas de Avalia√ß√£o de Classifica√ß√£o
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# ================================
# 1) Definir matriz de confus√£o (valores arbitr√°rios)
# ================================
VP = 50   # Verdadeiros Positivos
VN = 40   # Verdadeiros Negativos
FP = 10   # Falsos Positivos
FN = 5    # Falsos Negativos

# ================================
# 2) F√≥rmulas das m√©tricas
# ================================
acuracia = (VP + VN) / (VP + VN + FP + FN)
precisao = VP / (VP + FP) if (VP + FP) != 0 else 0
sensibilidade = VP / (VP + FN) if (VP + FN) != 0 else 0  # Recall
especificidade = VN / (VN + FP) if (VN + FP) != 0 else 0
f1_score = (2 * precisao * sensibilidade) / (precisao + sensibilidade) if (precisao + sensibilidade) != 0 else 0

# ================================
# 3) Exibir resultados em tabela
# ================================
resultados = {
    "Acur√°cia": round(acuracia, 3),
    "Precis√£o": round(precisao, 3),
    "Sensibilidade (Recall)": round(sensibilidade, 3),
    "Especificidade": round(especificidade, 3),
    "F1-Score": round(f1_score, 3)
}

df_resultados = pd.DataFrame(resultados, index=["Valor"])
display(df_resultados)

# ================================
# 4) Plotar Matriz de Confus√£o
# ================================
matriz = np.array([[VP, FN],
                   [FP, VN]])

labels = np.array([["VP", "FN"],
                   ["FP", "VN"]])

plt.figure(figsize=(6,5))
sns.heatmap(matriz, annot=labels, fmt="", cmap="Blues", cbar=False,
            xticklabels=["Classe Positiva", "Classe Negativa"],
            yticklabels=["Classe Positiva", "Classe Negativa"])

plt.title("Matriz de Confus√£o", fontsize=14)
plt.show()

# ================================
# 5) Explica√ß√£o Did√°tica
# ================================
print("üìå Explica√ß√£o das M√©tricas:\n")
print("üîπ Acur√°cia: Propor√ß√£o de previs√µes corretas (VP + VN) / Total.")
print("üîπ Precis√£o: Dos exemplos previstos como positivos, quantos realmente s√£o positivos? (VP / (VP + FP))")
print("üîπ Sensibilidade (Recall): Dos exemplos que eram positivos, quantos foram detectados? (VP / (VP + FN))")
print("üîπ Especificidade: Dos exemplos que eram negativos, quantos foram detectados corretamente? (VN / (VN + FP))")
print("üîπ F1-Score: M√©dia harm√¥nica entre precis√£o e recall. Boa para dados desbalanceados.")
```
