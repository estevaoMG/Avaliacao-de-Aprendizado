```python
# =========================================
# 📊 Cálculo de Métricas de Avaliação de Classificação
# =========================================

import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd

# ================================
# 1) Definir matriz de confusão (valores arbitrários)
# ================================
VP = 50   # Verdadeiros Positivos
VN = 40   # Verdadeiros Negativos
FP = 10   # Falsos Positivos
FN = 5    # Falsos Negativos

# ================================
# 2) Fórmulas das métricas
# ================================
acuracia = (VP + VN) / (VP + VN + FP + FN)
precisao = VP / (VP + FP) if (VP + FP) != 0 else 0
sensibilidade = VP / (VP + FN) if (VP + FN) != 0 else 0  # Recall
especificidade = VN / (VN + FP) if (VN + FP) != 0 else 0
f1_score = (2 * precisao * sensibilidade) / (precisao + sensibilidade) if (precisao + sensibilidade) != 0 else 0

# ================================
# 3) Exibir resultados em tabela
# ================================
resultados = {
    "Acurácia": round(acuracia, 3),
    "Precisão": round(precisao, 3),
    "Sensibilidade (Recall)": round(sensibilidade, 3),
    "Especificidade": round(especificidade, 3),
    "F1-Score": round(f1_score, 3)
}

df_resultados = pd.DataFrame(resultados, index=["Valor"])
display(df_resultados)

# ================================
# 4) Plotar Matriz de Confusão
# ================================
matriz = np.array([[VP, FN],
                   [FP, VN]])

labels = np.array([["VP", "FN"],
                   ["FP", "VN"]])

plt.figure(figsize=(6,5))
sns.heatmap(matriz, annot=labels, fmt="", cmap="Blues", cbar=False,
            xticklabels=["Classe Positiva", "Classe Negativa"],
            yticklabels=["Classe Positiva", "Classe Negativa"])

plt.title("Matriz de Confusão", fontsize=14)
plt.show()

# ================================
# 5) Explicação Didática
# ================================
print("📌 Explicação das Métricas:\n")
print("🔹 Acurácia: Proporção de previsões corretas (VP + VN) / Total.")
print("🔹 Precisão: Dos exemplos previstos como positivos, quantos realmente são positivos? (VP / (VP + FP))")
print("🔹 Sensibilidade (Recall): Dos exemplos que eram positivos, quantos foram detectados? (VP / (VP + FN))")
print("🔹 Especificidade: Dos exemplos que eram negativos, quantos foram detectados corretamente? (VN / (VN + FP))")
print("🔹 F1-Score: Média harmônica entre precisão e recall. Boa para dados desbalanceados.")
```
